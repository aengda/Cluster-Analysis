```python
import pandas

import pylab as pl

from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

```python
variables = pandas.read_csv('Matrix.csv')
Y = variables[['health_score']]
X = variables[['dailypct']]
```

```python
Nc = range(1, 20)

kmeans = [KMeans(n_clusters=i) for i in Nc]

kmeans

score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]

score

pl.plot(Nc,score)

pl.xlabel('Number of Clusters')

pl.ylabel('Score')

pl.title('Elbow Curve')

pl.show()
```

```python
pca = PCA(n_components=1).fit(Y)

pca_d = pca.transform(Y)

pca_c = pca.transform(X)
```

```python
kmeans=KMeans(n_clusters=5)

kmeansoutput=kmeans.fit(Y)

kmeansoutput

pl.figure('5 Cluster K-Means')

pl.scatter(pca_c[:, 0], pca_d[:, 0], c=kmeansoutput.labels_)

pl.xlabel('dailypct')

pl.ylabel('Health Score')

pl.title('5 Cluster K-Means')

pl.show()
```

```python
import pandas

import pylab as pl

from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

```python
Data = pandas.read_csv('HealthScore.csv')
X = Data[['zipcode']]
Y = Data[['health_score']]
```

```python
Data = pandas.read_csv('HealthScore.csv')
Xs = Data.iloc[:,0]
Ys = Data.iloc[:,1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(Xs,Ys)
plt.show()
```

```python
from sklearn.cluster import KMeans 
model=KMeans(n_clusters=5)
model.fit(Data)
##KMeans(algorithm='autho', ...)
labels = model.predict(Data)
```

```python
print(labels)
```

```python
plt.scatter(Xs,Ys, c=labels, alpha=0.5)
centroids = model.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
plt.scatter(centroids_x,centroids_y, marker='D', s = 50)
plt.show()
```
